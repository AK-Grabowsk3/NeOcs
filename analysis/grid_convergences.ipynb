Time grid convergence

```python
time_steps_au = [50, 100, 200, 400, 600, 1000, 2000, 4000, 6000]
xlabel = "Time step [au]"

xpi0, xpi10, xpi11, bsigma0, bsigma10, bsigma11 = get_reaction_rate_dependence("time_step", time_steps_au, identity=200)

fig, ax = plot_reaction_rate_dependence_0(xlabel, time_steps_au, xpi0, bsigma0)
ax.set_xscale("log")
fig, ax = plot_reaction_rate_dependence_1(xlabel, time_steps_au, xpi10, xpi11, bsigma10, bsigma11)
ax.set_xscale("log")
fig, ax = plot_ratio_dependence(xlabel, time_steps_au, xpi0, xpi10, xpi11, bsigma0, bsigma10, bsigma11)
ax.set_xscale("log")
```
