```python
import numpy
import matplotlib.pyplot as plt
import numpy.polynomial.legendre as leg
import utility
from potential import *

r_start = 50.0 / 1024.0
r_end = 50.0
r_points_no = 1024
r_step = (r_end - r_start) / (r_points_no - 1)
r_grid = np.array([r_start + i * r_step for i in range(r_points_no)])

polar_points_no = 160
x, _ = leg.leggauss(polar_points_no)
theta_grid = np.flip(np.arccos(x))

path = "../potentials"
```

Intermolecular potential

```python
V_grid = potential_to_npy(path, "potential.dat", r_grid, theta_grid, kx=5, ky=5)

fig, ax = utility.plot()
for i in range(0, len(theta_grid), 32):
    ax.plot(r_grid, V_grid[:, i] * cm_inv, label=r"$\theta = {:.2f}$".format(theta_grid[i]))
ax.legend()
ax.set_xlim(0, 20)
ax.set_ylim(-100, 900)
ax.set_xlabel(r"Distance [a$_o$]")
ax.set_ylabel(r"Potential [cm$^{-1}$]")
plt.show(ax)


np.clip(V_grid, -5e-4, 2e-3, out=V_grid)
fig, ax = plt.subplots()
cs = ax.contourf(theta_grid, r_grid, V_grid * cm_inv, cmap="hot", levels=100)
ax.set_ylim(0, 20)

ax.set_xlabel(r"Angle [rad]")
ax.set_ylabel(r"Distance [a$_o$]")
cbar = fig.colorbar(cs)
cbar.ax.set_ylabel(r"Potential [cm$^{-1}$]")
```

XPi Sigma

```python
V_grid = potential_to_npy(path, "XPi_gamma.dat", r_grid, theta_grid, lambda x: np.sqrt(x), lambda x: x**2, kx=5)

fig, ax = plt.subplots()
for i in range(0, len(theta_grid), 32):
    ax.plot(r_grid, V_grid[:, i] * cm_inv, label=r"$\theta = {:.2f}$".format(theta_grid[i]))
ax.legend()
ax.set_xlim(0, 20)
ax.set_ylim(-100, 1500)
ax.set_xlabel(r"Distance [a$_o$]")
ax.set_ylabel(r"X$\Pi$ gamma [cm$^{-1}$]")

np.clip(V_grid, -5e-4, 2e-3, out=V_grid)
fig, ax = plt.subplots()
cs = ax.contourf(theta_grid, r_grid, V_grid * cm_inv, cmap="hot", levels=100)
ax.set_ylim(0, 20)


ax.set_xlabel(r"Angle [rad]")
ax.set_ylabel(r"Distance [a$_o$]")
cbar = fig.colorbar(cs)
cbar.ax.set_ylabel(r"X$\Pi$ gamma [cm$^{-1}$]")
```
